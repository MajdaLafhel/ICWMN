```python
#1  -    save this code    -   extract list of categories
import csv

dict_categories = dict()
i =0


import pandas as pd

# Load the data from a CSV file
data_set = pd.read_csv('/content/network_metadata.csv', sep=';')

# Extract the third column of the DataFrame
rows = data_set.iloc[:, 2].values
# print(rows)
for row in rows:
  categories = str(row).split('-')
  for category in categories:
    if category not in dict_categories:
      dict_categories[category] = i
      i+=1

print(dict_categories)
len(dict_categories)
```

```python
#2  -   save this code    -   export categories and movies in a csv files

dict_movies_categories = dict()

with open('/content/network_metadata.csv', 'r') as file:
    reader = csv.reader(file)

    for index, row in data_set.iterrows():
      cat = []
      # print(row[0])
      movie = row[0]
      categ = row[2]

      categories = str(categ).split('-')

      for category in categories:
        cat.append(dict_categories[category])

      dict_movies_categories[movie] = cat

print(dict_movies_categories)

#------------------------------------------------

import csv

# Create a new CSV file
with open('movie_categories2.csv', 'w', newline='') as csvfile:
  writer = csv.writer(csvfile, delimiter=',')

  # Write the header row
  writer.writerow(['movie', 'categories'])

  # Iterate through the dictionary and write each key-value pair to a new row
  for key, value in dict_movies_categories.items():
    writer.writerow([key, value])
```

```python
#3  - Merge csv of network features + csv of categories in 1 csv
#export each feature in a separated vector

import json
import csv
import os

reversed_dict = {value: key for key, value in dict_categories.items()}

data_dict = dict()
# print(reversed_dict)
# print(dict_categories)

import pandas as pd
import ast
import numpy as np
df1 = pd.read_csv('/content/drive/MyDrive/gexf/movie_features2.csv')
# df1 = pd.read_csv('/content/drive/MyDrive/gexf/Normalized_movie_features2.csv')


df2 = pd.read_csv('/content/drive/MyDrive/gexf/movie_categories2.csv')

df2['categories']= df2['categories'].apply(ast.literal_eval)
df2['categories'] = df2['categories'].apply(lambda x: x[0])

# print(df2)

merged_df = pd.merge(df1, df2, on='movie')

merged_df['target'] = merged_df['categories'].map(reversed_dict)

merged_df.to_csv('/content/drive/MyDrive/gexf/movie_featureVector_categories2.csv', index=False, sep=';')

print(merged_df)
```
